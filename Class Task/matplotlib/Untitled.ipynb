```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=np.arange(1,11)
y=np.arange(11,21)
```

```python
plt.plot(x,y,c="g")
plt.xlabel("X axis")
plt .ylabel("Y axis")
plt.title("numpy array")
 
```

```python
plt.scatter(x,y,c="b")
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,"go")
plt.subplot(2,2,2)
plt.plot(x,y,"r--")
plt.subplot(2,2,3)
plt.scatter(x,y)
plt.subplot(2,2,4)
plt.plot(x,y,"r*",linestyle="dotted", linewidth=1,markersize=5)
plt.savefig("plot.png")
         
```

```python
x=np.arange(0,4*np.pi)
y=np.sin(x)
plt.title("Sine Wave")
plt.plot(x,y)
```

```python
x=np.arange(0,5*np.pi)
y=np.cos(x)
plt.title("Sine Wave")
plt.plot(x,y)
```

```python
x=np.arange(0,5*np.pi)
y_sin=np.sin(x)
y_cos=np.cos(x)
plt.title(" Wave")
plt.subplot(2,1,1)
plt.plot(x,y_cos, "go")
plt.xlabel("x")
plt.ylabel("x-cos")
plt.subplot(2,1,2)
plt.plot(x,y_sin, "r--")
plt.xlabel("x")
plt.ylabel("x-sin")
```
